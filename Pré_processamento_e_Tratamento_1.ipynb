```
import pandas as pd
import statistics as sts
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.preprocessing import LabelEncoder


caminho = '/content/drive/My Drive/microdados_enem_2019/MICRODADOS_ENEM_2019_PE.csv'
dados_pe = pd.read_csv(caminho)
```

```
dados_pe.head()
```

```
# Encontrar index da feature
dados_pe.columns.get_loc('NU_NOTA_REDACAO')
```

**Substitui√ß√£o dos valores nulos pela mediana**

```
# Cria√ß√£o e inicio de tratamento 
# Nota em Matematica
mediana = dados_pe['NU_NOTA_MT'].median()
dados_pe['NU_NOTA_MT'].fillna(mediana, inplace=True)
```

```
# Ciencias da natureza
mediana = dados_pe['NU_NOTA_CN'].median()
dados_pe['NU_NOTA_CN'].fillna(mediana, inplace=True)
```

```
# Ciencias Humanas
mediana = dados_pe['NU_NOTA_CH'].median()
dados_pe['NU_NOTA_CH'].fillna(mediana, inplace=True)
```

```
# Linguagens e seus c√≥digos
mediana = dados_pe['NU_NOTA_LC'].median()
dados_pe['NU_NOTA_LC'].fillna(mediana, inplace=True)
```

```
mediana = dados_pe['NU_NOTA_REDACAO'].median()
dados_pe['NU_NOTA_REDACAO'].fillna(mediana, inplace=True)
```

**Cria√ß√£o da coluna M√©dia**

```
notas = dados_pe.iloc[:,  np.r_[91:95,110]].values
notas
```

```
media = []

for i in notas:
  media.append(sts.mean(i))
media
```

```
categoria_media = []
for i in media:
  if (i <=200):
    categoria_media.append('E')
  elif (i>200 and i <=400):
    categoria_media.append('D')
  elif (i>400 and i <=600):
    categoria_media.append('C')
  elif (i>600 and i <=800):
    categoria_media.append('B')
  else:
    categoria_media.append('A')
categoria_media
```

```
dados_pe['CATEGORIA_MEDIA'] = categoria_media
```

```
dados_pe['NU_MEDIA'] = media
```

# Verificando a influ√™ncia dos dados socioecon√¥micos mais importantes do participante na media das notas.

```
# Verifica√ß√£o de dados duplicados
dados_pe[dados_pe.duplicated(['NU_INSCRICAO'],keep=False)]
```

```
dados_pe.columns.get_loc('CATEGORIA_MEDIA')
```

```
previsores = dados_pe.iloc[:, 111:136].values
classe = dados_pe.iloc[:, 136].values
classe
```

```
labelencoder = LabelEncoder()
for i in range(4):
  previsores[:,i] = labelencoder.fit_transform(previsores[:, i])
print(previsores[0])
```

```
for i in range(5,25):
  previsores[:,i] = labelencoder.fit_transform(previsores[:, i])
print(previsores[0])
```

```
x_treino, x_teste, y_treino, y_teste = train_test_split(previsores,
                                                        classe,
                                                        test_size = 0.3,
                                                        random_state = 123)
```

```
floresta = RandomForestClassifier(n_estimators=20)
floresta.fit(x_treino,y_treino)
```

```
importancia = floresta.feature_importances_
importancia
```

Podemos observar que os atributos mais importantes se encontram entre Q001 e Q006

```
previsao = floresta.predict(x_teste)
acerto = accuracy_score(previsao,y_teste)
acerto
```
